```python
import pulp
```

```python
demanda_2m = 50
demanda_3m = 60
demanda_4m = 90
```

```python
padroes = {
    "P1": [3, 0, 0],
    "P2": [1, 1, 0],
    "P3": [0, 2, 0],
    "P4": [1, 0, 1],
    "P5": [0, 0, 1],
    "P6": [0, 1, 1],
}
```

```python
padroes.pop("P6")
```

```python
modelo = pulp.LpProblem("Problema_Corte_de_Barras", pulp.LpMinimize)
```

```python
x = pulp.LpVariable.dicts("UsoPadrao", padroes.keys(), lowBound=0, cat=pulp.LpInteger)
```

```python
modelo += pulp.lpSum([x[p] for p in padroes.keys()]), "Minimizar_Barras"
```

```python
modelo += pulp.lpSum([padroes[p][0] * x[p] for p in padroes.keys()]) >= demanda_2m
```

```python
modelo += pulp.lpSum([padroes[p][1] * x[p] for p in padroes.keys()]) >= demanda_3m
```

```python
modelo += pulp.lpSum([padroes[p][2] * x[p] for p in padroes.keys()]) >= demanda_4m
```

```python
modelo.solve(pulp.PULP_CBC_CMD(msg=True))
```

```python
print("\nStatus da solução:", pulp.LpStatus[modelo.status])
print("Quantidade mínima de barras utilizadas:", pulp.value(modelo.objective))
```

```python
print("\nPlano de corte sugerido:\n")

for p in padroes.keys():
    if x[p].value() > 0:
        print(f"{p}: usar {int(x[p].value())} barras")
```
